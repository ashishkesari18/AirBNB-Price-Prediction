## Possible potential challenges:

1. Too Many Properties: In crowded places like New York City, there might be too many properties available. This means landlords and sellers have to compete more, which can bring down how much money they make from rent or selling.

2. Money Problems: Sometimes, if the economy isn't doing well or if there are money issues in a certain area, people might not want to rent or buy properties there. This can make the rent or property prices go down, so investors don't make as much money.

3. Rules and Costs: Sometimes, the government might make strict rules or charge high taxes on properties. This can eat into the money landlords or sellers make, making the investment less profitable.

4. Not Enough Good Stuff Around: If there aren't enough good things like shops, parks, or transportation options nearby, people might not want to live in an area. This can make it harder for landlords to find tenants, which means they don't make as much money.

5. People Moving Away or Changing: Sometimes, if the people living in an area change or if there are more older people and fewer younger people, it can affect how much demand there is for housing. If fewer people want to live there, landlords might not make as much money.

# Linear Regression

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)

y_pred_lr = lr.predict(x_test)
```

```python
mae_lr = metrics.mean_absolute_error(y_test, y_pred_lr)
mse_lr = metrics.mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_lr))
r2_lr = metrics.r2_score(y_test, y_pred_lr)


print('\nMean Absolute Error of Linear Regression     : ', mae_lr)
print('\nMean Squarred Error of Linear Regression     : ', mse_lr)
print('\nRoot Mean Squarred Error of Linear Regression: ', rmse_lr)
print('\nR2 Score of Linear Regression                : ', r2_lr)
```

1. Absolute  Error: The absolute error of the linear regression model is about 0.37. This means, on average, the model's predictions are off by around 0.37 units from the actual values.

2. Accuracy: The mean squared error, which measures the average of the squared errors, is about 0.25. Lower values indicate better accuracy, so the model's predictions are relatively close to the actual values.

3. Performance: The root mean squared error, which is the square root of the mean squared error, is about 0.50. This metric gives a sense of the typical size of the errors in the model's predictions. A lower value is better, indicating that the model's predictions are closer to the actual values.

4. Prediction Strength: The R2 score of the linear regression model is about 0.51. This score represents how well the model explains the variance in the data. An R2 score closer to 1 indicates a strong model that can explain a large portion of the variance in the data. In this case, the model explains about 51% of the variance, suggesting moderate predictive strength.
